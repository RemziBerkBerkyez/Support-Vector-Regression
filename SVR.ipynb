```python
AudiData['model'] = pd.Categorical(AudiData['model'])
dfDummies3 = pd.get_dummies(AudiData['model'], prefix = 'model')
dfDummies3
```

```python
AudiData=pd.concat([AudiData, dfDummies3], axis=1)
```

```python
AudiData=AudiData.drop("model",axis=1)
```

```python
y=AudiData["price"].values
x=AudiData.drop("price",axis=1).values
```

```python
y = y.reshape(-1,1)
```

```python
from sklearn.preprocessing import MinMaxScaler
target=MinMaxScaler()
predictor = MinMaxScaler()
x_ =predictor.fit_transform(x)
y_ = target.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x_,y_,test_size=0.25,random_state=42)
```

```python
from sklearn.svm import SVR
```

```python
svr_model = SVR(kernel = "rbf").fit(x_train, y_train)
```

```python
y_predict = svr_model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
np.sqrt(mean_squared_error(y_test, y_predict))
```

```python
tahmin = target.inverse_transform(y_predict.reshape(-1,1))

gerçek = target.inverse_transform(y_test.reshape(-1,1))
```

```python
from sklearn.metrics import mean_absolute_error as mae
  
# calculate MAE
error = mae(tahmin, gerçek)
  
# display
print("Mean absolute error : " + str(error))
```

```python
from sklearn.model_selection import GridSearchCV
svr_parameters = {"C": [0.0001,0.001,0.01,0.1,1,5], "gamma": [0.0001,0.001,0.01,0.1,1]}

svr_cv_model = GridSearchCV(svr_model, svr_parameters, cv = 10, n_jobs= 1,verbose=3).fit(x_train, y_train)
```

```python
pd.Series(svr_cv_model.best_params_)
```

```python
svr_tuned = SVR("rbf", C = (5), gamma = (0.01)).fit(x_train, y_train)
```

```python
y_predict = svr_tuned.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```
